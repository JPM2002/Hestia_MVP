```python
import os
import pandas as pd
import numpy as np

# üìÅ Directory containing your CSVs
csv_dir = "NittanyBusinessDataset"

# üîç List all CSV files in the directory
csv_files = [f for f in os.listdir(csv_dir) if f.endswith(".csv")]

# ‚úÖ Print options
print("Available CSV files:")
for i, file in enumerate(csv_files, 1):
    print(f"{i}. {file}")

# üì• Ask user for selection
try:
    choice = int(input("Enter the number of the file you want to preview: "))
    selected_file = csv_files[choice - 1]
except (ValueError, IndexError):
    print("Invalid selection. Exiting.")
    exit()

# üìÑ Full path to the selected CSV
csv_path = os.path.join(csv_dir, selected_file)

# ‚úÖ Load and display
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    df = df.replace(r'^\s*$', np.nan, regex=True)  # Clean empty cells
    
    # üõ†Ô∏è Show all columns, rows, and full cell content
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    pd.set_option('display.width', 1000)
    pd.set_option('display.max_colwidth', None)
    
    print(f"\n--- Preview of {selected_file} ---\n")
    print(df.head(20).to_string(index=False))  # üëà full preview, no truncation
else:
    print(f"File not found: {csv_path}")

```
